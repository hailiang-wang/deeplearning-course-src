## 张量连续性质 contiguous

张量连续：就是从一个元素访问下一个元素的时候，不需要跳跃。[If we move along a single row in this tensor, the elements are placed as in its storage and we do not have to jump over any number. A tensor like this is called “contiguous”. ](https://medium.com/swlh/deep-learning-with-pytorch-tensor-basics-part-1-stride-offset-contiguous-tensors-5d87476b7d9f)

https://medium.com/swlh/deep-learning-with-pytorch-tensor-basics-part-1-stride-offset-contiguous-tensors-5d87476b7d9f

https://stackoverflow.com/questions/48915810/what-does-contiguous-do-in-pytorch

* 使用张量的 `is_contiguous()` 接口，会返回一个张量是否连续。
* 使用张量的 `contiguous()` 接口，会将一个张量变成需连续的。
* 进行 narrow(), view(), expand(), transpose() 等操作要求 Tensor 是连续的。

# 索引 indexing
Tensor 的索引类似于 Python List 的索引和分片。

比如一个 AxBxC 的三个维度的 Tensor a。

a[第0维的分片, 第1维的分片, 第2维的分片]

分片的语法和 Python List 分片语法一致，开始:结束:步进。

更多索引的[高级语法介绍](https://medium.com/@AI-Simplified/mastering-pytorch-indexing-simple-techniques-with-practical-examples-154c232b6d84)。

```python
print("*" * 8, " a")
a = torch.randn(5,4,3)
print(a)

print("*" * 8, " b")
b = a[1,]     # 只要第 0 维的第一个成员
print(b)

print("*" * 8, " c")
c = a[1:]   # 第 0 维从第一个成员开始都要，注意：这里索引从 0 开始
print(c)

print("*" * 8, " d")
d = a[1:, 1] # 第 0 维从第一个成员开始都要，第二维只要第一个成员
print(d)
```

## 广播机制

在 Tensor 的加减运算中，当两个 tensor 不能直接符合数学的运算规则时，PyTorch 会先尝试将 tensor 进行变换，再进行计算，这个变换的规则就是：broadcasting。

1. 通过适当复制元素来扩展一个或两个数组，以便在转换之后，两个张量具有相同的形状；
2. 对生成的数组执行按元素操作。

![](../../assets/book/screenshot_20260415115755.png)

**在大多数情况下，我们将沿着数组中长度为1的轴进行广播**，如下例子：



```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

由于`a`和`b`分别是$3\times1$和$1\times2$矩阵，如果让它们相加，它们的形状不匹配。
我们将两个矩阵*广播*为一个更大的$3\times2$矩阵，如下所示：矩阵`a`将复制列，
矩阵`b`将复制行，然后再按元素相加。




```python
a + b
```

## 乘法运算

torch.tensor 有四种乘法运算，分别是：*, torch.mul, torch.mm, torch.matmul.

### 点乘( * 或 torch.mul )

a与b做*乘法，原则是如果a与b的 size 不同，则以某种方式将a或b进行复制，使得复制后的a和b的size相同，然后再将a和b做 **按元素运算（element-wise）** 的乘法。




```python
a = torch.randn(3,4)
b = torch.randn(3,4)
c = a * b
print(c)

d = torch.mul(a,b)
print(d)

```

如果是一个矩阵 和 一个标量相乘，则是，每个矩阵的元素都和该标量相乘得到新的 矩阵。

如果是一个矩阵 A 和一个向量 b 相乘，那么，进行广播操作。
但是要求，向量的列，和矩阵的列是一致的。
广播操作的原理是，此时是将向量复制多行，得到矩阵 B，复制后和矩阵 A 形状一样。

```python
a = torch.ones(3,4)
b = torch.Tensor([1,2,3,4])


c = a * b
print(c.shape, a.shape)

print(c)
print("*" * 8, a)
```

如果是一个矩阵，和另外一个矩阵相乘。

那么，也是看形状是否一致，不一致，再进行广播，运算。
广播后，也可能无法运算，则会报错。

```python
a = torch.randn(3,4,1)
b = torch.randn(4,1)
c = a*b
print(c)
print(c.shape)
```

### 标准矩阵乘法（torch.mm）

数学里的矩阵乘法，要求两个Tensor的维度满足矩阵乘法的要求，即第一个矩阵的列，必须等于第二个矩阵的行。

* 不支持广播
* 只接受矩阵，向量；不支持二维以上输入

```python
a = torch.ones(3,4)
b = torch.ones(4,2)
c = torch.mm(a,b)
print(c)
print(c.shape)
```

### 支持广播的矩阵乘法（torch.matmul）

torch.matmul 是 PyTorch 中用于执行矩阵乘法的函数，它根据输入张量的维度自动选择适当的矩阵乘法方式，包括：

* 向量内积（1D @ 1D）
* 矩阵乘向量（2D @ 1D）
* 向量乘矩阵（1D @ 2D）
* 矩阵乘矩阵（2D @ 2D）
* 批量矩阵乘法（>2D）

torch.mm的高级版本，支持多维，自动 broadcast。


```python
a = torch.randn(3,4)
b = torch.randn(5,4,2)
c = torch.matmul(a, b)
print(c)
print(c.shape)
```
